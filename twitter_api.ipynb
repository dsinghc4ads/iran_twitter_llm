```python
import requests
import pandas as pd

import time

def get_user_id(username, bearer_token):
    url = f"https://api.twitter.com/2/users/by/username/{username}"
    headers = {"Authorization": f"Bearer {bearer_token}"}
    response = requests.get(url, headers=headers)
    return response.json()["data"]["id"]

def get_tweets(user_id, bearer_token):
    url = f"https://api.twitter.com/2/users/{user_id}/tweets"
    headers = {"Authorization": f"Bearer {bearer_token}"}
    params = {"max_results": 100}  # Adjust this value to retrieve more tweets
    response = requests.get(url, headers=headers, params=params)
    return response.json()

# Replace 'your_bearer_token' with your actual Bearer Token
# BEARER_TOKEN = "your_bearer_token"

user_list = ['PadDolat', 
             'Iran_GOV', 
             'khamenei_ir', 
             'PressTV', 
             'HS_Soleimani', 
             'Tasnimnews_Fa', 
             'khabaronlinee', 
             'Khabar_F', 
             'FarsNews_Agency',
             'mehrnews_ir',
             'MehrnewsCom',
             'isna_farsi',
             'IRNA_1313', 
             'IrnaEnglish']

def create_dataset(tweets, user, dataset):
    print(tweets)
    for tweet in tweets['data']:
        dataset['text'].append(tweet['text'])
        dataset['user'].append(user)
    return dataset

dataset = {'text':[], 'user':[]}
for user in user_list[6:9]:
    user_id = get_user_id(user, BEARER_TOKEN)
    tweets = get_tweets(user_id, BEARER_TOKEN)
    dataset = create_dataset(tweets, user, dataset)
    #time.sleep(900)


dataset = pd.DataFrame(dataset)
dataset.to_csv('tweets-3.csv')
```
